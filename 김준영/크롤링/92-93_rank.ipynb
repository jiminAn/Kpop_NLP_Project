```python
idol= pd.read_csv('k-pop_92-93.tsv',sep='\t')
```

```python
idol
```

```python
idol.columns = ['년도','가수','제목','남녀','장르','작사가','작곡가','소속사','가사']
```

```python
idol
```

```python
idol=idol.reset_index().rename(columns={"index": "최고 순위"})
```

```python
idol
```

```python
idol=idol[["년도","가수","제목","남녀","장르","최고 순위","작사가","작곡가","소속사","가사"]]
```

```python
idol
```

```python
idol.loc[idol["남녀"] == "남","남녀"] = 'male'
idol.loc[idol["남녀"] == "여","남녀"] = 'female'
idol.loc[idol["남녀"] == "혼","남녀"] = 'mixed'
```

```python
idol
```

```python
idol=idol.dropna(subset=['가사'])
```

```python
idol
```

```python
idol.to_csv('k-pop_92-93_r.tsv',index=False,header=None,sep='\t')
```
