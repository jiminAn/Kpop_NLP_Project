```python
idol_list = pd.read_csv('92to99_idol_list.csv', encoding = 'utf-8')
idol_list.shape
```

```python
artists = idol_list.artist
print(artists)
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(r'C:\Users\Eunhoo\Code\NLP\chromedriver.exe')
driver.get("https://www.melon.com/search/total/index.htm?q=%EC%86%8C%EB%85%80%EC%8B%9C%EB%8C%80&section=&searchGnbYn=Y&kkoSpl=N&kkoDpType=&ipath=srch_form")
```

```python
artist_id = []
```

```python
# n=0
tmp = []
id = ""
for artist in artists:
    sleep(3)
    search = driver.find_element_by_xpath('//*[@id="top_search"]')
    search.send_keys(Keys.CONTROL + "a");
    search.clear()
    #WebDriverWait(driver, timeout)
    search.send_keys(artist)

    # 아티스트명 검색
    search.send_keys(Keys.RETURN)

    try:
        
        # 아티스트 클릭
        driver.find_element_by_xpath('//*[@id="conts"]/div[3]/div/div[1]/div/a/strong').click()
        url = driver.current_url
        tmp = url.split('=')
        id = tmp[1]
        print(artist,id)
        artist_id.append(id)

    
    except NoSuchElementException:
        artist_id.append('')
```

```python
idol_id = artist_id.copy()
```

```python
idol_id = pd.DataFrame(idol_id)
idol_id.head()
```

```python
idol_id.columns = ['idol_id']
idol_id.head()
```
